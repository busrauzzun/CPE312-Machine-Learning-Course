```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("mushrooms.csv")
```

```python
sns.histplot(dataset['class']);
```

```python
x = dataset.drop(['class'],axis=1)
```

```python
y = dataset['class']
```

```python
x = pd.get_dummies(x)
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
y = enc.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state =1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
model_gini = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
model_gini.fit(x_train,y_train)
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(model_gini.fit(x_train,y_train));
```

```python
y_pred = model_gini.predict(x_test)
```

```python
y_pred_train = model_gini.predict(x_train)
```

```python
accuracy_score_train = accuracy_score(y_train,y_pred_train)
accuracy_score_train
```

```python
accuracy_score_test = accuracy_score(y_test,y_pred)
accuracy_score_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(cm, annot=True, linewidth=0.5, linecolor='red',fmt=".0f",ax=ax)
plt.show()
```

```python
report = classification_report(y_test,y_pred)
```
